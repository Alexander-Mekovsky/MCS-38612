```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
california_counties = pd.read_csv("california_counties.tsv", sep="\t", usecols=range(1, 9))
california_counties
```

```python
covid_cases = pd.read_csv("covid_surge_offsets.tsv", sep="\t", usecols=range(1, 6))
covid_cases
```

```python
merged = pd.merge(california_counties, covid_cases, left_on='name', right_on='area')
merged.drop('area', axis=1, inplace=True)
merged
```

```python
merged['population_density'] = (merged['population_2020']/merged['area_sq_miles'])
merged
```

```python
sns.scatterplot(merged, x='surge_4_offset', y='population_density', style='general_law_or_charter', hue='general_law_or_charter', palette='pastel')
sns.set_palette("deep")
plt.title('Surge 4 Offset of Log Population Density')
plt.yscale('log')
plt.ylabel('Log Population Density')
plt.xlabel('Surge 4 Offset')
plt.legend(title='General Law or Charter')
plt.show()
```
