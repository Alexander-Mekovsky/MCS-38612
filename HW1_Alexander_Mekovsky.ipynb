```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

#Todo: Remove California from area column, idealy by removing any values marked by 'state'
```

```python
df = pd.read_csv('covid19cases_test.csv')
```

```python
df_Cleaned = df.dropna()
```

```python
df_Cleaned
```

```python
df_Cleaned.isnull().sum()
```

```python
df_Cleaned.loc[:, 'date'] = pd.to_datetime(df_Cleaned['date']).dt.date
reference_date = datetime.strptime('2020-02-01', '%Y-%m-%d').date()
df_Cleaned.loc[:, 'date_offset'] = df_Cleaned['date'].apply(lambda x: (x - reference_date).days)
```

```python
print(df_Cleaned)
```

```python
df_Unique_Area = df_Cleaned.drop_duplicates(subset='area', keep='first')
```

```python
df_Unique_Area
```

```python
df_Unique_Area = df_Unique_Area.drop(df_Unique_Area.index[5])
population = df_Unique_Area['population'].sum()
print(population)
```

```python
df_Cleaned.loc[:, 'normalized_cases'] = df_Cleaned['cases']/population
```

```python
elDorado = df_Cleaned[df_Cleaned['area'] == 'El Dorado']
elDorado
```

```python
fig, axes = plt.subplots()
axes.plot(elDorado['date_offset'], elDorado['normalized_cases'])
```

```python
safe_cols_df = df_Cleaned.drop(columns=['date', 'area', 'area_type'])
medians_df = safe_cols_df.groupby(['date_offset']).median().reset_index()
```

```python
medians_df
```

```python
first_quarter_df = safe_cols_df.groupby(['date_offset']).quantile(0.25).reset_index()
third_quarter_df = safe_cols_df.groupby(['date_offset']).quantile(0.75).reset_index()
```

```python
first_quarter_df
```

```python
third_quarter_df
```

```python
medians_df = medians_df.rolling(window=7).mean()
first_quarter_df = first_quarter_df.rolling(window=7).mean()
third_quarter_df = third_quarter_df.rolling(window=7).mean()
```

```python
fig, axes = plt.subplots(figsize=(10,8))

axes.plot(first_quarter_df['date_offset'], first_quarter_df['normalized_cases'], label='Q1')
axes.plot(medians_df['date_offset'], medians_df['normalized_cases'], label='Median')
axes.plot(third_quarter_df['date_offset'], third_quarter_df['normalized_cases'], label='Q3')

axes.fill_between(first_quarter_df['date_offset'], 
                  first_quarter_df['normalized_cases'], 
                  third_quarter_df['normalized_cases'], 
                  color='gray', alpha=0.2, label='Area Between Q1 and Q3')

axes.set_title('California Coronavirus Cases')
axes.set_xlabel('Time (Days)')
axes.set_ylabel('Normalized Cases')
axes.legend()

plt.show()
```
